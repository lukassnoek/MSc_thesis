### Simulating data: One subject
Because I don't have appropriate data for this analysis at hand, I'll create a random (gaussian noise) dataset.
I need to set some (non-sense) parameters to initialize the useful mvpa_mat object.
Here, I assume that our data consists of 60 trials with two factors:

factor_let, with levels 'A' and 'B' 

factor_num, with levels '1' and '2'


```python
data = np.random.normal(10, 5, (60, 100))                   # data of 60 trials by 100 features
subject_name = 'test'
mask_name = 'random_gauss'                                  # gaussian noise
mask_index = None
mask_shape = None
factor_let = ['a'] * 30 + ['b'] * 30                        # factor 1
factor_num = [1] * 15 + [2] * 15 + [1] * 15 + [2] * 15      # factor 2
class_labels = zip(factor_let, factor_num)                  # zip factors (list of tuples)
```

### Initializing 'mvpa_mat' object
Here, I'll initialize the variable 'test' as an instance of the 'mvpa_mat' class.
This class has a couple of nice attributes, that come in handy later in the analyses.

```python
test = mvpa_mat(data, subject_name, mask_name, mask_index, mask_shape, class_labels)
```

### From mvpa_mat to RDM
Here, a Representational Dissimilarity Matrix (RDM) is created by taking the
Euclidian distance between all trial, pair-wise.
It is actually a fancy name for a distance matrix.

```python
RDM = ME.create_RDM(test)
plt.imshow(RDM)
plt.colorbar()
```

### Creating RDM predictors
Now, we need to create predictors for the 'observed RDM' above. 
Here, we assume that the distance between trials belonging to the same class is 0 (they are encoded similarly),
while the distance between trials of different classes is 1 (they are encoded differently).
A predictor is made for each factor, and their interaction term.


```python
predictors = ME.create_rdm_regressor(test)
```
