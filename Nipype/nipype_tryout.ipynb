```python
"""
Start of a (f)MRI processing pipeline using the nipy/nipype environment:
nipy.org
Gorgolewski K., et al. (2011). Front. Neurinform. 5:13
...
Lukas Snoek, fall 2014
"""

# To do:
# Physiological noise regressors in FSL >> write script using FSL's PNM
# Implement call to matlab-script of Steven's MSTC-function
# Build in warnings/errors
# Use absolute paths in glob.glob >> funcDirs: USE os.path.abspath()

# Notes to Self:
# If there is no output, check the directory! (print os.getwd())


# Importing necessary packages
import nipype.interfaces.fsl as fsl
import nipype.interfaces.io as nio
import nipype.pipeline.engine as pe
import nipype.interfaces.utility as util
import os
import glob
import shutil
import warnings
```

```python
# Specification of pipeline-general variables
homeDir = '/media/lukas/Data/Sample_fMRI' 
subject_stem = 'HWW_'
study_stem = 'PM'
func_names = [['HWW', 'blocked_pictures.log', 'pictures.phy'],
              ['Zinnen1','eventrelated_sentences_1','zinnen1.phy'],
              ['Zinnen2','eventrelated_sentences_2','zinnen2.phy']]

# Physiological files should have the extension .phy (not .log, bc of confusion with presentation-files)
```

```python
"""
Converts par/rec files to nii.gz files and creates separate directories for each file within subject directories.
Moves par/rec files to backup directory. Also moves .log and .phy files.
"""

os.chdir(homeDir)

# Convert all par/recs to nii.gz
os.system("parrec2nii *.PAR")
os.system("gzip *.nii")
for file in glob.glob('*.nii'):
    os.remove(file)

# Create back-up dir with all original par/recs
PARREC_backup = homeDir + '/PARREC_backup'
if os.path.exists(PARREC_backup) == 0:
    os.mkdir(PARREC_backup)        

parFiles = glob.glob('*.PAR')
recFiles = glob.glob('*.REC')

for file in parFiles:
    shutil.move(file, PARREC_backup)

for file in recFiles:
    shutil.move(file, PARREC_backup)

ToProcess = os.getcwd() + '/ToProcess'    

if os.path.exists(ToProcess) == 0:
    os.mkdir(ToProcess)    

# Move all files to ToProcess dir    
for file in glob.glob(subject_stem + '*'):
    shutil.move(file, ToProcess)

# Make dir for bash-files (BET bashfile-etc)
Log_dir = os.getcwd() + '/Logs'
if os.path.exists(Log_dir) == 0:
    os.mkdir(Log_dir)
        
os.chdir(ToProcess)
```

```python
# Make subject-directories. NB: YOU NEED TO HAVE A T1-FILE in your subjectDir.
subjects = glob.glob('*T1*.nii.gz')
subjectDir = []

for idx, subs in enumerate(subjects):
    
    subjectDir.append(subs[0:len(subject_stem) + 3])
    
    if os.path.exists(subjectDir[idx]) == 0:
        os.mkdir(subjectDir[idx])
    
    toMove = glob.glob(subs[0:len(subject_stem) + 3] + '*')
    
    # Move all files to subject-specific dirs
    for mFile in toMove:
        if os.path.isfile(mFile):
            shutil.move(mFile, subjectDir[idx])
            
unallocated = 0
for entry in glob.glob('*'):
    if os.path.isfile(entry):
        unallocated =+ 1

if unallocated > 0:
    warnings.warn('There are %s unallocated files in ToProcess' % unallocated)

subject_dir_list = [os.getcwd() + '/' + sDir for sDir in subjectDir]
            
# Make an mri-dir for each .nii.gz file & move nii.gz/phy/log files to that dir
T1_dir_list = []
func_dir_list = []

for subDir in glob.glob(subject_stem + '*'):
    if not os.path.isfile(subDir):
        os.chdir(subDir)
    else:
        warnings.warn('There are unallocated files in your subject directories')
    
    T1_dir = 'T1'
    if os.path.exists(T1_dir) == 0:
        os.mkdir(T1_dir)
    
    T1_dir_list.append(os.getcwd() + '/T1')
    
    T1_file = glob.glob(subject_stem + '*' + 'T1' + '*' + 'nii.gz')
    for t in T1_file:
        shutil.move(t, T1_dir)
    
    for mriDir in func_names:
        funcDir = mriDir[0]
        
        if os.path.exists(funcDir) == 0:
            os.mkdir(funcDir)
        
        func_dir_list.append(os.getcwd() + '/' + funcDir)
        
        funcFile = glob.glob(subject_stem + '*' + mriDir[0] + '*' + 'nii.gz')
        for f in funcFile:
            shutil.move(f, funcDir)
        
        logFile = glob.glob(subject_stem + '*' + mriDir[1] + '*')
        for l in logFile:
            shutil.move(l, funcDir)
        
        phyFile = glob.glob(subject_stem + '*' + mriDir[2] + '*')
        for p in phyFile:
            shutil.move(p, funcDir)
        
    os.chdir(ToProcess)    
```

```python
# Start nipype
os.chdir('/media/lukas/Data/Sample_fMRI/ToProcess/HWW_007/T1/')
skullstrip = fsl.BET()
skullstrip.inputs.in_file = '/media/lukas/Data/Sample_fMRI/ToProcess/HWW_007/T1/HWW_007_WIP_sT13DTFE_P2_S2_3m_SENSE_3_1.nii.gz'
skullstrip.inputs.terminal_output = 'none'
skullstrip.inputs.frac = 0.3

bet_results = skullstrip.run()
smooth = fsl.IsotropicSmooth(in_file=bet_results.outputs.out_file, fwhm=4)
smooth.run()
```
